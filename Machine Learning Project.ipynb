### data Correlation with Churn value:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
rs = np.random.RandomState(0)

df2 = d1
#Convertin the predictor variable in a binary numeric variable
df2['Churn'].replace(to_replace='Yes', value=1, inplace=True)
df2['Churn'].replace(to_replace='No',  value=0, inplace=True)

df_dummies = pd.get_dummies(df2)

plt.figure(figsize=(15,8))
df_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

## Normalization :


```python

X= df_dummies.drop(columns=['Churn'])
y= pd.factorize(df_dummies['Churn'])[0]
error = []
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

## KNN :
# best K choice


```python
from sklearn.neighbors import KNeighborsClassifier
for i in range(1, 44):
    knn = KNeighborsClassifier(i)
    knn_model = knn.fit(X_train, y_train)
    pred_i = knn_model.predict(X_test)
    error.append(np.mean(pred_i != y_test))
plt.figure(figsize=(12, 6))
plt.plot(range(1, 44), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Taux Erreur pour les differentes valeurs de k')
plt.xlabel('K ')
plt.ylabel('Erreur')
plt.savefig("KNN.png")
```


# with GridSearshCV:


```python
param_grid = {'n_neighbors' : np.arange(5,15),
              'metric' : ['euclidean', 'manhattan','minkowski'],
              'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']}
grid=GridSearchCV(KNeighborsClassifier(),param_grid,cv=6)
grid.fit(X_train, y_train)
print(grid.best_score_)
grid.best_params_
```

```python
from sklearn.metrics import plot_confusion_matrix,roc_curve,roc_auc_score, classification_report,auc,precision_recall_fscore_support
all_scores=pd.DataFrame()
def model_PR_RE_F1_ROC(model,X_train,y_train,X_test, y_test,name):
    y_pred = model.predict(X_test)
    precision, recall, f1, support = precision_recall_fscore_support(y_test, y_pred,average='binary', pos_label=0)
    scores = pd.DataFrame(columns=['Precision','Recall','F1','ROC'])
    scores.loc["methode 1:"+name] = [precision, recall, f1,roc_auc_score(y_test, y_pred)]
    scores
    return scores
    
                       
        
```

```python
T1=grid.best_params_
knn = KNeighborsClassifier(n_neighbors=T1['n_neighbors']  ,metric=T1['metric'] ,algorithm=T1['algorithm'])
knn_model = knn.fit(X_train, y_train)
y_pred_knn =knn_model.predict(X_test)
print('train score = ', knn_model.score(X_train, y_train))
print('test_score = ', knn_model.score(X_test, y_test))
Res=model_PR_RE_F1_ROC(knn_model,X_train,y_train,X_test, y_test,'knn')
all_scores = all_scores.append(Res)

```

```python
def correlatedFeatures(dataset, threshold):
    correlated_columns = set()
    correlations = dataset.corr()
    for i in range(len(correlations)):
        for j in range(i):
            if abs(correlations.iloc[i,j]) > threshold:
                correlated_columns.add(correlations.columns[i])
    return correlated_columns

```
